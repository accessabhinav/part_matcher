```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
import numpy as np
import pandas as pd
import re

stopwords = set(stopwords.words("english"))

def listtostring(clean_line):
    return(" ".join(clean_line))

def replacer(string1):
    string2 = string1.partition(",")[0].partition(";")[0].partition("(")[0].partition("[")[0].partition('for')[0].partition('with')[0].partition('offer')[0]
    return(string2.replace('\bscr\b','screw').replace('\bspr\b','spring').replace('\bbd\b','band').replace('spr bd','spring band')
           .replace('\bw/h\b','wiring harness').replace('brg','bearing').replace('bkt','bracket').replace('\bwh\b','wiring harness')
           .replace('w/h','wiring harness').replace('\breinf\b','reinforcement').replace('\bwh\b','wiring harness').replace('mtg.','mounting')
           .replace('mtg','mounting').replace('\bbrkt\b','bracket').replace('\brr\b','rear').replace('frt','front').replace('reinfrocement','reinforcement').replace('mountingbracket','mounting bracket'))
    
#using Lemmatization
lemma = nltk.wordnet.WordNetLemmatizer()

x_doc = []
y_doc = []
test_doc = []
test_line = []

data=pd.read_csv("E:\JS\Py\Text_similarity\Part_type_map.csv", sep=',')

clean_doc = []

for line in data['DESCRIPTION']:
    clean_line = [word for word in line.split() if word not in stopwords]
    
    for word in line.split():
        test_line.append(lemma.lemmatize(word))
            
    #test_doc.append(listtostring(test_line))
    test_line = listtostring(test_line)
    test_line1 = replacer(test_line.lower())
    test_doc.append(test_line1)
    test_line = []
    
#clean_doc.append(listtostring(clean_line))
x_doc = test_doc
    
y_doc = data['Part_Type']

text = '. '.join([str(item) for item in x_doc])

#generate bigrams and trigrams

tokens = nltk.word_tokenize(text)
words = [word for word in tokens if word.isalnum() and len(word)>2 and word!='lh' and word!='rh']

#Create your bigrams
bigram = list(nltk.bigrams(words))

#Create your trigrams
trigram = list(nltk.trigrams(words))

#n-gram rationalization

bigram_list = []
for line in bigram:
    if(line[0]==line[1]):
        bigram_list.append(str(line[0]))
    else:
        bigram_list.append(str(line[0])+" "+str(line[1]))
    
trigram_list = []
for line in trigram:
    if(line[0]==line[1] and line[0]==line[2]):
        trigram_list.append(str(line[0]))
        
    if(line[0]==line[1] and line[0]!=line[2]):
        trigram_list.append(str(line[0])+" "+str(line[2]))
        
    if(line[0]!=line[1] and line[0]==line[2]):
        trigram_list.append(str(line[0])+" "+str(line[1]))
    
    if(line[1]==line[2] and line[1]!=line[0]):
        trigram_list.append(str(line[0])+" "+str(line[1]))
        
    if(line[1]==line[0] and line[1]!=line[2]):
        trigram_list.append(str(line[0])+" "+str(line[2]))
        
    if(line[2]==line[0] and line[2]!=line[1]):
        trigram_list.append(str(line[1])+" "+str(line[2]))
    
    if(line[0]!=line[1] and line[0]!=line[2]):
        trigram_list.append(str(line[0])+" "+str(line[1])+" "+str(line[2]))


trigram_list = list(set(trigram_list))
bigram_list = list(set(bigram_list))

trigram_list.sort(key=len, reverse = True)
bigram_list.sort(key=len, reverse = True)

#new_bigrams = [b for b in bigram_list if all(b not in t for t in trigram_list)]

#compute frequency distribution for all the bigrams in the text
fdist = nltk.FreqDist(bigram)

for line in x_doc:
    found_flag = 0
    
    print("\n",line," is part type:")

    for item in trigram_list:
        if(line.find(item)!=-1):
            print(item)
            found_flag = 1

    for item in bigram_list:
        if(found_flag==0 and line.find(item)!=-1):
            print(item)
            
    found_flag = 0
                    

# for line in x_doc:
#     found_flag = 0
    
#     print("\n",line," is part type:")

#     for item in trigram_list:
#         if(line.find(item)!=-1):
#             found_flag = 1
#             print(item)

#         if(found_flag==1):
#             break
            
#         else:
#             for item in bigram_list:
#                 if(line.find(item)!=-1):
#                     print(item)
#                     break
#         break
```
